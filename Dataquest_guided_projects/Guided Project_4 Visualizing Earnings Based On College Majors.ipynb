```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')

print (recent_grads.iloc[0])
print (recent_grads.head())
print (recent_grads.tail())
print (recent_grads.describe())

raw_data_count = len(recent_grads)

recent_grads = recent_grads.dropna()

clean_data_count = len(recent_grads)

print (raw_data_count)
print (clean_data_count)
```

```python
samplesize_median = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
samplesize_median.set_title('Sample Size and Median')

samplesize_unemp = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
samplesize_unemp.set_title('Sample Size and Unemployment Rate')

fulltime_median = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
samplesize_median.set_title('Employed Full Time and Median')

sharedwomen_unemp = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
sharedwomen_unemp.set_title('Women as Share of Total and Unemployment Rate')

men_median = recent_grads.plot(x='Men', y='Median', kind='scatter')
men_median.set_title('Men and Median')

women_median = recent_grads.plot(x='Women', y='Median', kind='scatter')
women_median.set_title('Women and Median')
```

```python
histcols = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']

fig = plt.figure(figsize=(5,12))
for i in range (1,5):
    ax = fig.add_subplot(8, 1, i)
    ax = recent_grads[histcols[i]].plot(kind='hist', rot=35)

fig2 = plt.figure(figsize=(5,12))
for i in range (4,8):
    ax = fig2.add_subplot(8, 1, i-3)
    ax = recent_grads[histcols[i]].plot(kind='hist', rot=35)
```

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```
